```python

import cv2
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r'C:/Program Files/Tesseract-OCR/tesseract.exe'
from google_trans_new import google_translator
from tkinter import *
from tkinter import filedialog 
translator=google_translator()

window = Tk() 
window.title('Convert image to text and different languages') 
window.geometry("1000x750")


def browseFiles():
    global im
    global text_comb
    global new
    global adaptive_threshold
    new=""
    filename = filedialog.askopenfilename(initialdir = "/", title = "Select a File",  filetypes = (("Image files","*.jpg*"),("all files", "*.*")))
    im=cv2.imread(filename)
    scale_percent=0.75
    width=int(im.shape[1]*scale_percent)
    height=int(im.shape[0]*scale_percent)
    dimension=(width,height)
    im=cv2.resize(im,dimension,interpolation=cv2.INTER_AREA)
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    adaptive_threshold = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 45, 20)
    text_comb=pytesseract.image_to_string(adaptive_threshold)
    leave = " to my his her at this us use of off your from in out a it  that the is am without are was were being been has what why who when with have had and one two three four five six seven eight nine ten be do did done can could may might shall should will would must ought"
    output=text_comb.split(' ')
    leave1=leave.split(' ')
    for i in output:
        i=i.lower()
        found = 1
        for j in leave1:
            if i==j:
                found = -1
        if found==1:
            new=new+" "+i                
    new=new.split(' ')
    button_view_image = Button(window,text="View image",command=viewImage).grid(row=3,column=0)
    button_display = Button(window,  
                        text = "View words in image", 
                        command = displayText).grid(row=3,column=1)
    button_convert_hi = Button(window,text= "Hindi ", command= convertHindi).grid(row=4,column=0)
    button_convert_pa = Button(window,text= "Punjabi", command= convertPunjabi).grid(row=4,column=1)
    button_convert_ta = Button(window,text= "Tamil", command= convertTamil).grid(row=5,column=0)
    button_convert_te = Button(window,text= "Telugu", command= convertTelugu).grid(row=5,column=1)
    
def convertHindi():
    my_output.delete(1.0,END)
    for k in new:
        word_hi=translator.translate(k, lang_src='en', lang_tgt='hi')
        wrd=k+"     "+word_hi+"\n"
        my_output.insert(END,wrd)
    
    
def convertPunjabi():
    my_output.delete(1.0,END)
    for k in new:
        word_pu=translator.translate(k, lang_src='en', lang_tgt='pa')
        wrd=k+"     "+word_pu+"\n"
        my_output.insert(END,wrd)
    

        
def convertTamil():
    my_output.delete(1.0,END)
    for k in new:
        word_ta=translator.translate(k, lang_src='en', lang_tgt='ta')
        wrd=k+"     "+word_ta+"\n"
        my_output.insert(END,wrd)
    

def convertTelugu():
    my_output.delete(1.0,END)
    for k in new:
        word_te=translator.translate(k, lang_src='en', lang_tgt='te')
        wrd=k+"     "+word_te+"\n"
        my_output.insert(END,wrd)
    
        

def displayText():
    my_output.delete(1.0,END)
    my_output.insert(END,text_comb)
    
def viewImage():
    
    cv2.imshow("window",im)
    
def Save():
    saveFile = filedialog.asksaveasfilename(defaultextension=".txt",title="Save Files", filetypes=[("Text Files",".txt")])
    if saveFile:
        saveFile=open(saveFile,'w', encoding='utf-8')
        saveFile.write(my_output.get(1.0,END))
        saveFile.close()
    else:
        lb=Label(window,text="Enter a File name to save").grid(row=5,column=2)
    
label_file_explorer = Label(window,  
                            text = "Extract text from your image", 
                            width=50, height = 10,  
                            fg = "Red")
label_open_image = Label(window,text="Click here to open image")

label_file_explorer.grid(row=0,column=0)
label_open_image.grid(row=1,column=0)
button_explore = Button(window,  
                        text = "Open image", 
                        command = browseFiles).grid(row=2,column=0)

my_output = Text(window,width=50,height=10)
my_output.grid(row=3,column=2)
save_button = Button(window,text="Save File",command=Save).grid(row=4,column=2)
Exit_button = Button(window,text="Click To EXIT", command= window.destroy).grid(row=6,column=2)  
 
window.mainloop() 
```
